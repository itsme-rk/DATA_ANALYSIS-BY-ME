#### started making the model let's split the dataset based on the values we need as the target column and the columns the we required for the trainning process

```python
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import  ColumnTransformer
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.tree import DecisionTreeClassifier
```

```python
data_cleaned.head()
```

```python
data_cleaned.drop(columns=['transaction_id','customer_id','merchant_id','transaction_time'],inplace=True)
```

```python
data_cleaned
```

```python
x_train,x_test,y_train,y_test = train_test_split(data_cleaned.drop(columns=['is_fraudulent']),data_cleaned['is_fraudulent'],test_size=0.2,
                                                random_state=42)
```

```python
x_train
```

```python
y_train
```

#### Lets encode the catagorical columns  and send it through pipeline to transform it into the  scaler or miin max 

```python
##One hot encoding
trf1 = ColumnTransformer([
    ('ohe_card_location_purchase_fraud',OneHotEncoder(sparse=False,handle_unknown='ignore'),['card_type','location','purchase_category','fraud_type'])
],remainder='passthrough')

trf1
```

```python
x_train['card_type'].nunique() 
```

```python
x_train['location'].nunique() 
```

```python
x_train['purchase_category'].nunique()
```

```python
x_train['fraud_type'].nunique()
```

```python
trf2 = ColumnTransformer([
    ('scaled_values',StandardScaler(),slice(0,7))
],remainder='passthrough')


trf2
```

### let's pass it throuhgh the pipeline

```python
trf4 = DecisionTreeClassifier(min_samples_split=10)
```

```python
pipe1 = Pipeline([
    ('ohe1', trf1),
    ('stdscaller', trf2),
    ('pred_model', trf4)
])
```

```python
pipe1
```

```python
# Display Pipeline

from sklearn import set_config
set_config(display='diagram')
```

```python
pipe1.fit(x_train,y_train)
```

```python
# Predict
y_pred = pipe1.predict(x_test) 
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```

```python
def make_prediction(input_data, pipeline):
    transformed_data = pipeline.named_steps['ohe1'].transform(input_data)
    
    predictions = pipeline.named_steps['pred_model'].predict(transformed_data)
    
    return predictions

new_input_data = pd.DataFrame({
    'amount': [100.0],
    'customer_age': [35],
    'card_type': ['Rupay'],
    'location': ['Surat'],
    'purchase_category': ['Digital'],
    'fraud_type': ['scam']
})

predictions = make_prediction(new_input_data, pipe)

print(predictions)
```
